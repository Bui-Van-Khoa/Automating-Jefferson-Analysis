```python
from pyannote.audio import Pipeline
pipeline = Pipeline.from_pretrained(
  "pyannote/speaker-diarization-3.0",
  use_auth_token=auth_token)

diarization = pipeline("twos.wav")


with open("audio.rttm", "w") as rttm:
    diarization.write_rttm(rttm)
```

```python
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"Speaker {speaker}: {turn.start:.1f}s - {turn.end:.1f}s")
```

```python
import csv

diarization2 = pipeline("twos.wav", num_speakers=2)

csv_filename = "diarization_results.csv"


with open(csv_filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Speaker", "Start Time (s)", "End Time (s)"])

    for turn, _, speaker in diarization2.itertracks(yield_label=True):
        writer.writerow([speaker, round(turn.start, 1), round(turn.end, 1)])
```
